```python
import pandas as pd

path_to_file = "C:/Users/souri/Desktop/tryco2/Data"

df = pd.read_csv(path_to_file + "/owid-co2-data.csv")

pd.set_option('display.max_columns', None)

df.sample(5)
```

```python
df.info()
```

```python
df = df.loc[:, ["year",
                "iso_code",
                "country",
                "energy_per_capita",
                "energy_per_gdp",
                "coal_co2",
                "oil_co2",
                "gas_co2",
                "cement_co2",
                "gdp",
                "population",
                "land_use_change_co2",
                "flaring_co2",
                "consumption_co2",
                "trade_co2_share",
                "other_industry_co2"
                ]]
df.sample(5)
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['population','gdp','coal_co2','oil_co2','gas_co2','cement_co2'], how='all')
```

```python
#missing for fuel and emmission often mean near 0 so replace by 0

fuel_cols = ['coal_co2','oil_co2','gas_co2','cement_co2','energy_per_capita','energy_per_gdp','land_use_change_co2','flaring_co2', "consumption_co2", "trade_co2_share", "other_industry_co2"]
df[fuel_cols] = df[fuel_cols].fillna(0)

df[['gdp','population']] = df[['gdp','population']].fillna(df[['gdp','population']].median())
```

```python
df["country"].unique()
```

```python
df_clean = df[df['iso_code'].str.fullmatch(r'[A-Z]{3}', na=False)].reset_index(drop=True)

print('Rows after filter:', len(df_clean))
print('Unique countries after filter:', df_clean['country'].nunique())
```

```python
# Create total CO2 column as sum of component sources
co2_components = [
    'coal_co2', 'oil_co2', 'gas_co2', 'cement_co2',
    'land_use_change_co2', 'flaring_co2', 'consumption_co2', 'other_industry_co2'
]

co2_fossils = [
    'coal_co2', 'oil_co2', 'gas_co2', 'cement_co2'
]

# fill missing components with 0 and sum
df_clean['co2_total'] = df_clean[co2_components].fillna(0).sum(axis=1)
df_clean['co2_fossils'] = df_clean[co2_fossils].fillna(0).sum(axis=1)
```

```python
min_year = 1950
max_year = 2022

df_clean = df_clean[(df_clean['year'] >= min_year) & (df_clean['year'] <= max_year)].reset_index(drop=True)
```

```python
df_sheet1.columns = df_sheet1.columns.str.strip()

df_sheet1 = df_sheet1.drop(columns=['Region', 'Sub-region', 'Producer Type', 'Sub-Technology', 'Technology', 'M49 code'], errors='ignore')

df_sheet1 = df_sheet1[df_sheet1['RE or Non-RE'] != 'Total Non-Renewable']

df_sheet1.sample(5)
```

```python
df_sheet1 = df_sheet1.rename(columns={"ISO3 code": "iso_code", "Year": "year"})

df_merged = pd.merge(df_sheet1, df_clean, on=['iso_code', 'year'], how='right')

df_merged.to_csv("../Data/clean_data.csv", index=False)
```
